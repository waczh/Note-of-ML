### ***3.1.2复习感知机***

感知机仅有两层神经元构成,输出只有0和1。判断条件为$b+w1X+w2X$是否大于0 $\Rightarrow$ $\Rightarrow$

b(偏置)可以被视为一个仅输出的神经元 $\Rightarrow$ $\Rightarrow$ 

根据上式($b+w1X+w2X$)的判断条件，可以将$y=h(x)$与$f(x)=b+w1X+w2X$复合$h(f(x))$$\longrightarrow$$y=h(b+w1X+w2X)$

此时的$h(x)$是阶跃激活函数(x指原先的$b+w1X+w2X$) $\Rightarrow$$\Rightarrow$

### ***3.1.3激活函数登场***

激活函数是将输入信号总和转化为输出信号的函数(activation function),输入信号总和是$b+w1X+w2X$,输出信号是$y=h(x)$，映射方式为$h()$

神经元内激活函数计算的过程$\Rightarrow$$\Rightarrow$

其中，单层网络是用阶跃激活函数，神经网络使用的是sigmoid激活函数

### ***3.2.6非线性函数***

非线性函数指的是函数表达式不为形如$f(x)=ax+b$的函数。阶跃函数和$sigmoid$函数均为非线性函数

线性函数不能作为神经网络的激活函数，神经网络中含有隐藏层，而等效叠加的隐藏层是没有意义的$\Rightarrow$$\Rightarrow$<br>假设激活函数$h(x)=ax+b$则$h(h(x))=k^2x+(kb+b)$相当于只有一层激活函数$f(x)=k^2x+(k+1)b$

### ***3.4 三层神经网络的实现***

三层神经网络包含三层权重和三层偏置，可使用numpy实现

### ***3.4.3  代码实现小结***

定义一个函数包含每一层对应的权重和偏置，和另一个函数实现传参和前进运算

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Mar 25 16:35:45 2022

@author: qingchenweifeng
"""
import numpy as np
def sigmoid(x):      #激活函数1
    return 1/(1+np.exp(-x))
def identity_func(x):#激活函数2
    return x
def inital_network():#初始化所有层的权重和偏置
    network={}#创建一个字典
    network['W1']=np.array([[0.1,0.3,0.5],[0.2,0.4,0.6]])
    network['W2']=np.array([[0.1,0.4],[0.2,0.5],[0.3,0.6]])
    network['W3']=np.array([[0.1,0.3],[0.2,0.4]])
    network['b1']=np.array([0.1,0.2,0.3])
    network['b2']=np.array([0.1,0.2])
    network['b3']=np.array([0.1,0.2])
    return network#将所有初始化的对象传入空字典
def forward_func(network,X1):  #X1是初始的输入信号
    W1,W2,W3=network['W1'],network['W2'],network['W3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    A1=np.dot(X1,W1)+b1
    Z1=sigmoid(A1)         #实现向前算法
    A2=np.dot(Z1,W2)+b2
    Z2=sigmoid(A2)
    A3=np.dot(Z2,W3)+b3
    return identity_func(A3)
network=inital_network()
X1=np.array([1.0,0.5])
Y=forward_func(network, X1)#传入初始化数据
print(Y)
```

### ***3.5  输出层的设计***

输出层的设计指的是设计输出层对应的激活函数，根据不同的实现目的分为**恒等函数**和**softmax**函数

**恒等函数**用于分类的问题上比如物体识别分类；**softmax**用于梯度下降等线性回归问题上

### ***3.5.1 恒等函数和softmax函数***

恒等函数即$\sigma$(),输出结果与输入完全相同$\Rightarrow$ $\sigma$(x)=$x$

$softmax$函数表达式$\Rightarrow$

$$ y_{k}=\frac {exp(a_{k})} {\sum_{i=1}^{n} {exp(a_{i})} } \quad $$

即第$k$个的输出等于第$k$个的e取指除所有输出的e取指的和，因此输出层的各个神经元都受每个输入的影响

$softmax$的代码实现$\Rightarrow$

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Mar 25 17:46:51 2022

@author: qingchenweifeng
"""
import numpy as np
def softmax(a): # a是整个输出数据构成的数组
    a_exp=np.exp(a)# 所有元素指数化
    a2_sum=np.sum(a_exp)#指数化后求和
    return a_exp/a2_sum
z=np.array([0.3,2.9,4.0])
print(softmax(z))
```

### ***3.5.3 实现softmax函数时的注意事项***

因为$softmax$的式子中$exp(a_{i})$=$e^{a_{i}}$当$a_{i}$过大时会造成溢出，导致结果不确定，因此我们可以恒等改写$softmax$ $\Rightarrow$

$$ y_{k}=\frac {exp(a_{k})}{\sum_{i=1}^{n}{exp(a_{i})} }\quad =\frac {exp(a_{k}+C')} {\sum_{i=1}^{n} {exp(a_{i}+C')} } \quad $$

改写后的代码实现$\Rightarrow$$\Rightarrow$

```python
# -*- coding: utf-8 -*-
"""
Created on Fri Mar 25 17:46:51 2022

@author: qingchenweifeng
"""
import numpy as np
def softmax(a):
    c=-np.max(a) #令c=-max(a)则可以避免数据溢出的情况
    a_exp=np.exp(a+c)
    a2_sum=np.sum(a_exp)
    return a_exp/a2_sum
z=np.array([1010,1000,900])
print(softmax(z))
```

### ***3.5.3 softmax函数的特征***

由$softmax$的函数表达式可以看出，对每个$y_{k}$进行求和得到结果为1，因此可以把$softmax$的输出结果看作是该输出为$ture$的可能性，常用于机器学习中

```python
import numpy as np
def softmax(a):
    c=-np.max(a)
    a_exp=np.exp(a+c)
    a2_sum=np.sum(a_exp)
    return a_exp/a2_sum
z=np.array([4,5,6])
print(softmax(z))
```

比如可以将输出的结果看作是“第一个类别的正确率是$9$%,第二个类别的正确率是$24.4$%,第三个类别的正确率是$66.5$%”

当机器学习后进行推理时，由于其$softmax$的单调性，激活前最大的值激活后不会被改变，因此可以忽略$softmax$,同时节约了计算机的算力

### ***3.5.4 输出层的神经元数量***

输出层$a_{1},a_{2}...a_{i}$的个数取决于待解决问题的对象的个数，比如识别数字$0-9$对应$10$个对象，从而输出层得有$10$个神经元

如机器识别数字，灰度（Gray scale）代表输出的大小$\Rightarrow$$\Rightarrow$

"2"对应的灰度最大，直接输出$y2$,推理过程中可不用$softmax$

### ***3.6.2 神经网络的推理处理***

输入层有28*28=784个神经元，在图像处理中每个输入有坐标索引和灰度值；输出层有10个神经元，是0—9中所有数的个数

正规化(nomoralization)是预处理(pre-processing)的一种，用代码表示是load_mnist()中的nomoralize=Ture,实现推理处理的过程$\Rightarrow$$\Rightarrow$

```python
# -*- coding: utf-8 -*-
"""
Created on Sat Mar 26 17:05:14 2022

@author: wangchen
"""
import numpy as np
import pickle #实现数据序列和反序列化
from dataset.mnist import load_mnist#从mnist.py脚本中导入load_mnist函数
def sigmoid(x):
    return 1/(1+np.exp(-x))#隐藏层的激活函数
def softmax(x):
    c=-np.max(x)
    y_i=np.exp(x+c)   #输出层的激活函数
    y_k_sum=np.sum(y_i)
    return y_i/y_k_sum
def get_data():   #返回测试对象和测试标签（真实值）
    (x_train,t_train),(x_test,t_test)=load_mnist(normalize=True,flatten=True,one_hot_label=False)#正规化、一维数组、原标签
    return x_test,t_test#获得测试对象和标签
def init_network():
    with open("sample_weight.pkl",'rb') as f:#读入学习到的权重和偏置
        network=pickle.load(f)
    return network
def predict(network,x):
    W1,W2,W3=network['W1'],network['W2'],network['W3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    A1=np.dot(x,W1)+b1
    Z1=sigmoid(A1)
    A2=np.dot(Z1,W2)+b2#推理过程：向前传播算法+学习后数据=预测值
    Z2=sigmoid(A2)
    A3=np.dot(Z2,W3)+b3
    Y=softmax(A3)
    return Y
x,t=get_data()#x为训练图像对应的灰度矩阵，t为图片对应的真实数字
network=init_network()#以字典的形式将学习后数据传给形参，参与推理过程
accuracy_count=0#精度初始化为0
for i in range(len(x)):
    y=predict(network, x[i])#传入灰度矩阵参与推理预测
    Y=np.argmax(y)#取最大值，即最大概率的对应元素的下标，为预测标签
    if Y==t[i]:#判断预测标签与真实标签是否一致
        accuracy_count+=1
print("Accuracy:"+str(float(accuracy_count)/len(x)))#识别精度=正确预测数/总数
```

### ***3.6.3批处理***

每一个经过load_mnist(）读入的x是一张图片的灰度值矩阵，且x[0].shape=(784,);若同时处理100个这样大小的灰度矩阵，矩阵仍满足相乘的条件

相比于逐步计算，处理数值计算的库能高效处理大型数组的运算，可以将更多的时间用在计算上

```python
# -*- coding: utf-8 -*-
"""
Created on Sat Mar 26 17:05:14 2022

@author: wangchen
"""
import numpy as np
import pickle
from dataset.mnist import load_mnist#从mnist.py脚本中导入load_mnist函数
def sigmoid(x):
    return 1/(1+np.exp(-x))
def softmax(x):
    c=-np.max(x)
    y_i=np.exp(x+c)
    y_k_sum=np.sum(y_i)
    return y_i/y_k_sum
def get_data():
    (x_train,t_train),(x_test,t_test)=load_mnist(normalize=True,flatten=True,one_hot_label=False)
    return x_test,t_test
def init_network():
    with open("sample_weight.pkl",'rb') as f:
        network=pickle.load(f)
    return network
def predict(network,x):
    W1,W2,W3=network['W1'],network['W2'],network['W3']
    b1,b2,b3=network['b1'],network['b2'],network['b3']
    A1=np.dot(x,W1)+b1
    Z1=sigmoid(A1)
    A2=np.dot(Z1,W2)+b2
    Z2=sigmoid(A2)
    A3=np.dot(Z2,W3)+b3
    Y=softmax(A3)
    return Y
x,t=get_data()
network=init_network()
batch_size=100
accuracy_count=0
for i in range(0,len(x),batch_size):#（始，末，步长），把总样本切片
    x_batch=x[i:i+batch_size]#从x个数据中取出第i个到第i+100个所有的数据
    y_batch=predict(network, x_batch)#100个批处理（预测值）
    p=np.argmax(y_batch,axis=1)#从行的方向取出每个图片的最大值，axis=1是矩阵的第一个维度
    accuracy_count+=np.sum(p==t[i:i+batch_size])#100个预测值和真实值比较后生成含有100元素的bool数组
print("Accuracy:"+str(float(accuracy_count)/len(x)))
```

### ***3.7小结***

• 神经网络中的激活函数使用平滑变化的sigmoid函数或ReLU函数。其中sigmoid函数是单调递增的

• 通过巧妙地使用NumPy多维数组，可以高效地实现神经网络。在向前传播中运用了numpy的点乘和求最大值的下标

• 机器学习的问题大体上可以分为回归问题和分类问题。两者的差异在于输出层的激活函数

• 关于输出层的激活函数，回归问题中一般用恒等函数，分类问题中一般用softmax函数。

• 分类问题中，输出层的神经元的数量设置为要分类的类别数。比如手写数字识别中10个数字对应十个输出层的神经元

• 输入数据的集合称为批。通过以批为单位进行推理处理，能够实现高速的运算。
